```python
import seaborn as sns 
import matplotlib.pyplot as plt 

# Distribution plote use for quantitative 
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

## Displote or histogram



```python
sns.displot(df['total_bill'],bins = 10 , kde = True)
```

### Joint plote
- compare two col

```python

sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python

sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python

sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

## pair plote

```python
sns.pairplot(df)
# use for usually analizing data 
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue = 'sex' , palette='rainbow')
```

## rug plote

```python
sns.rugplot(df['total_bill'])
```
