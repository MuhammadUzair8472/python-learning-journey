```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## play with catagorical data 

```python
# load data set tip
df = sns.load_dataset('tips')
df
```

```python
# we draw for information count plote 

sns.countplot( x = df['sex'] , hue= df['smoker'])
```

### Bar plote
* bar plote take x axis and y axis 

```python

# x axis for  categorical data and y axis for quantitative data now put it 
sns.barplot(x = df['sex'] , y= df['total_bill'])
```

```python
import numpy as np 
sns.barplot(x = df['sex'] , y =df ['total_bill' ], estimator = np.sum)
```

### Box plote

```python
sns.boxenplot(x = df['tip'] , y = df['day'] , data = df )
```

### violinplote

```python
sns.violinplot(x = 'day', y= 'total_bill',data = df)
```

```python
sns.stripplot(x='day', y='total_bill', data=df)
```
