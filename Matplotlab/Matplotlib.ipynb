```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

### matplot lab basic

```python
x = np.linspace(0 , 5 , 11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title('Hello')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### Subplos in matplote lab

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
```

Object Oriented Way

```python
fig = plt.figure ()
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)

axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

### Types Of Plots

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

### saving plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basic_plot")
```
